# Komparasi

Oke sekarang kita masuk ke latihan komparasi setelah kemarin latihan operator dalam matematika, sekarang kita lanjut masuk ke latihan selanjutnya yaitu komparasi. 

Apasih komparasi tuh? jadi kaya ngecek gitu doang sih apakah suatu nilai lebih besar/lebih kecil dsb dari suatu nilai yang lainnya dalam bentuk Boolean.

Semisal ada :
a = 5
b = 10

- apakah a > b ? **False** ya
- lalu, apakah b > a ? **True** ya.

gitu singkatnya.

Di python, untuk komparasi ada banyak :
- \> (lebih besar dari)
- \< (lebih kecil dari)
- \>= (lebih dari sama dengan)
- \<= (kurang dari sama dengan)
- == (sama dengan)
- != (tidak sama dengan)
- is
- is not

---

kita buat variable dulu deh

````python
a = 4
b = 2
c = 3
````

## > (Lebih besar dari) 

````python
print(f"apakah {a} > 3 ? {a > 3}")
````

`True` ya, karena variable a memiliki nilai value 4 yang dimana 4 pastinya lebih besar dari 3. 

````python
print(f"apakah {b} > 3 ? {b > 3}")
````

nah kalo kita mencoba membandingkan angka yang sama coba

````python
print(f"apakah {c} > 3 ? {c > 3}")
````

`False` ya, karena 3 itu seharusnya sama dengan 3. kalo pengen hasilnya `True`, harus lebih besar dari 3.

---

## < (lebih kecil dari)

kita copy paste aja ya code diatas dan kita ganti `>` menjadi `<`

````python
print(f"apakah {a} < 3 ? {a < 3}")
````

````python
print(f"apakah {b} < 3 ? {b < 3}")
````

````python
print(f"apakah {c} < 3 ? {c < 3}")
````

## >= (lebih dari sama dengan)

nah kalo >= atau <= semisal angkanya sama, hasilnya bisa `True`. kita cek satu-satu deh

````python
print(f"apakah {a} >= 3 ? {a >= 3}")
````

````python
print(f"apakah {b} >= 3 ? {b >= 3}")
````

````python
print(f"apakah {c} >= 3 ? {c >= 3}")
````

Di sebelumnya kan selalu salah kan baik `<` atau `>`. Nah disini kita menggunakan `>=` hasilnya bisa `True`.

kita cek deh :

````python
3.00000000000001 > 3
````

sedangkan kalo ada `>=`

````python
3 >= 3
````

## <= (kurang dari sama dengan)

````python
print(f"apakah {a} <= 3 ? {a <= 3}")
````

````python
print(f"apakah {b} <= 3 ? {b <= 3}")
````

````python
print(f"apakah {c} <= 3 ? {c <= 3}")
````

## == (sama dengan)

Ingat, untuk assignment menggunakan 1 `=`, sedangkan untuk komparasi menggunakan 2 `==` :

*variable assignment (menggunakan 1 `=`)*
```python
a = 4  # tanda = hanya satu
```

*komparasi*
```python
a == 4  # tanda == ada 2 dan menghasilkan boolean
```

````python
a == 4
````

````python
print(f"apakah {a} == 3 ? {a == 3}")
````

````python
print(f"apakah {b} == 3 ? {b == 3}")
````

````python
print(f"apakah {c} == 3 ? {c == 3}")
````

hasilnya baru `True` kalo angkanya benar-benar sama

---

## != (tidak sama dengan)

ya sesuai namanya, apakah *nilai_a* tidak sama dengan *nilai_b*? kalo tidak sama mengeluarkan hasil `True`

````python
a != 4
````

````python
print(f"apakah {a} != 3 ? {a != 3}")
````

````python
print(f"apakah {b} != 3 ? {b != 3}")
````

````python
print(f"apakah {c} != 3 ? {c != 3}")
````

kita lihat sebuah magic nih, kita buat variable baru

````python
x = 5
y = 5
````

````python
print(f"nilai x : {x}, id = {hex(id(x))}")
print(f"nilai x : {x}, id = {hex(id(x))}")
print(x is y)
````

*kenapa `hex id` nya bisa sama?* 

Di Python jika suatu variable memiliki nilai yang sama, akan ditaruh dalam memory yang sama. apalagi kalo tipe datanya sama. 

Berbeda dari bahasa pemograman lain walaupun nilai dan tipe datanya sama, akan disimpan tempat yang berbeda. (C++)

## is not

Secara penjelasan harusnya tidak jauh berbeda dengan `is`. 

````python
print(f"nilai x : {x}, id = {hex(id(x))}")
print(f"nilai x : {x}, id = {hex(id(x))}")
print(x is not y)
````
