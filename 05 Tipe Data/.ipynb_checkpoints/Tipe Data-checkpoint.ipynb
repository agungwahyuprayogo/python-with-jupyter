## Integer 
Angka satuan yang ga ada koma-nya

Oke, yang pertama adalah tipe data angka. **Angka disini adalah satuan, jadi tidak ada koma**. Nah kalo misalkan kita buat kaya gini :

````python
data_integer = 1 # angka satuan boleh 1, 4, 5, dst bahkan sampai sejuta pun boleh, asal tidak ada koma di angka tersebut
````

Nah, karena di **Python itu ngga ada yang kaya gini** :

```python
int data_interger = 1
```

kaya di bahasa pemograman lain, **dia akan otomatis meng-asign bahwa data interger** (angka satuan seperti `1`, `2`, dst) **adalah data interger.**

*"oke bang, coba diliatin gimana caranya?"* (cara liat tipe data)

oke, kita buat kaya gini :

```python
print(type(data_integer))
```

untuk melihat tipe datanya

````python
print("data :", data_integer, ", bertipe data ", type(data_integer))
````

oke, terlihat kan? kalo **variable `data_interger` bertipe data `int`**. terlihat dari hasil print `<class 'int'>`

## Float 
Angka dengan koma

Selanjutnya, kita punya tipe data kedua. Tipe data ini disebutnya sebagai *decimal* ya, atau *float*. **Float itu, adalah angka dengan koma.**

Jadi kita bisa buat disini :

````python
data_float = 1.5 # tanda koma menggunakan tanda titik
````

*Untuk melihat tipe datanya menggunakan Python, kita bisa cek lagi dengan cara diatas* dan **mengganti nama variable-nya** 

````python
print("data :", data_float, ", bertipe data ", type(data_float))
````

lihat? tipe **`data_float` dengan nilai yang kita masukan `1.5` bertipe data float**, terlihat dari `<class 'float'>`

## String

Selanjutnya adalah tipe data `String`. *"String itu apa?"* **String adalah kumpulan karakter.**

````python
data_string = "Agung" # gunakan tanda kutip
````

**Jangan lupa untuk menambahkan tanda kutip** -> ' ' **atau double kutip** -> " " . Sebelum dan sesudah kumpulan karakter yang kita masukan.

Nah kalo kita cek lagi, nanti hasilnya adalah sebagai berikut :

````python
print(f"data_string : {data_string}, bertipe data : {type(data_string)} ")
````

Oke, **`data_string` hasilnya adalah bertipe data _String_** karena dari hasil output terlihat hasilnya adalah `<class 'str'>`. 'str' itu adalah string.

*Bang kalo kita ubah dari `Agung` jadi `Agung 10` gimana?

````python
data_string = "Agung 10" # gunakan tanda kutip
print(f"data_string : {data_string}, bertipe data : {type(data_string)} ")
````

Jadi, **selama masih ada tanda kutip nya nilai di variable tersebut akan selalu di baca sebagai string**.

## Boolean
Tipe data biner (0,1) atau output True / False.

kita langsung cek aja ya

````python
data_bool = True
````

````python
print(f"data_bool : {data_bool}, bertipe data : {type(data_bool)}")
````

nah hasilnya sudah terlihat : `<class 'bool'>`. 

*Apasih 'bool' itu?* 

**Bool itu maksudnya adalah boolean.**

coba kita ganti dari `True` menjadi `False`

````python
data_bool = False
print(f"data_bool : {data_bool}, bertipe data : {type(data_bool)}")
````

sama hasilnya boolean

# Tipe Data Khusus

Oke, selanjutnya adalah tipe data khusus. Misalnya kalian anak matematik, ada beberapa tipe data yang dapat membantu kalian.

## Complex
**Bilangan Kompleks, ini adalah tipe data bilangan kompleks** seperti `10+5i`. 

````python
data_complex = complex(5, 6) # angka real di depan (5) imajiner di belakang (6)
print(f"data_complex : {data_complex}, bertipe data : {type(data_complex)}")
````

oke, terlihat ya hasilnya `(5+6j)` dimana `j` disini adalah *Imaginary*.

## Tipe data dari bahasa C

**Karena bahasa pemograman Python itu dibuatnya make bahasa C**. kita bisa juga menggunakan tipe data dari bahasa C. 

*"Tapi bagaimana caranya, bang?"* 

Caranya kita **import dulu si packaging atau library-nya**. Ada banyak sebenernya, ada `c_long`, `c_long_long` , dst. 

Kita gunakan salah satunya aja 

### c_doubel

````python
from ctypes import c_double
````

jadi, `c_double` ini bisa menampung data yang **lebih panjang dari float**. Anggap saja kita terpaksa import package ini ketika sudah amat jago dan dibutuhkan. 

````python
data_c_double = c_double(10.6) # jangan lupa ketik c_double dulu
````

````python
print(f"data_c_type : {data_c_double}, bertipe data {type(data_c_double)}")
````

oke, terlihat kan? hasilnya bukan float melainkan `<class 'ctypes.c_double'>`
