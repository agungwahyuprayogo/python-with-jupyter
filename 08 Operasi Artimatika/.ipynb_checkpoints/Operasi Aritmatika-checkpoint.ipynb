# Operasi Aritmatika

Sekarang kita masuk ke pelajaran operasi aritmatika. Jadi yang namanya operasi dalam pemograman tidak akan jauh dari operasi aritmatika seperti : 
- Penjumalah (+)
- Pengurangan (-)

- Pembagian (/)

- Modulus

kita buat dulu deh variable awal :

```python
a = 10
b = 3
c = -5
d = 1
```

## Operasi Tambah (+)

cukup simple ya, cukup gunakan tanda `+` di keyboard :

```python
print(f"hasil dari {a} + {b} = {a+b}")
```

```python
print(f"hasil dari {a} + {c} = {a+c}")
```

```python
print(f"hasil dari {c} + {d} = {c+d}")
```

Jangan bingung ya, anggap saja punya hutang 5 (-5) lalu baru bisa membayar 1. Maka sisa hutangnya adalah -4

## Operasi Pengurangan (-)

## Operasi Perkalian (*)

Bukan menggunakan `x` , melainkan menggunakan tanda bintang (`*`)

```python
print(f"hasil dari {c} x {d} = {c*d}")
```

```python
print(f"hasil dari {a} x {b} = {a*b}")
```

```python
print(f"hasil dari {a} x {c} = {a*c}")
```

Kalo bingung, cara ngapalinnya gini :

- plus dikali plus hasilnya plus
- plus dikali minus hasilnya minus
- minus dikali minus hasilnya minus
- minus dikali minus hasilnya plus

kita coba satu-satu deh kalo ga percaya

```python
print(3*3)       # plus kali plus
print(3*(-3))    # plus kali minus
print((-3)*3)    # minus kali plus
print((-3)*(-3)) # minus kali minus
```

## Operasi Pembagian (/)

Bukan menggunakan `:` di keyboard, melainkan menggunakan `/`

```python
print(f"hasil dari {a} : {b} = {a/b}")
```

Kita tahu hasil dari 10 : 3 adalah 3,3... Nah, bagusnya python adalah, dia akan otomatis menghasilkan angka dalam bentuk tipe `float`, jadi kita tidak perlu repot-repot mengubah tipe data terlebih dahulu.

```python
print(f"hasil dari {c} : {d} = {c/d}")
```

```python
print(f"hasil dari {a} : {c} = {a/c}")
```

Nah, kita tes apakah pembagian mirip-mirip kaya perkalian apa ga

```python
print(3/3)       # plus dibagi plus
print(3/(-3))    # plus dibagi minus
print((-3)/3)    # minus dibagi plus
print((-3)/(-3)) # minus dibagi minus
```

hmm mungkin ga terlalu keliatan ya karena angkanya sama, kita ganti deh angkanya

```python
print(10/3)       # plus dibagi plus
print(10/(-3))    # plus dibagi minus
print((-10)/3)    # minus dibagi plus
print((-10)/(-3)) # minus dibagi minus
```

## Operasi Eksponen / Pangkat (**)

jadi kita menggunakan 2 bintang -> `**`

```python
print(f"hasil dari {a}^{b} = {a**b}")
```

jangan bingungg... 10 x 10 = `100` x 10 = `1000`

```python
print(f"hasil dari {a}^{c} = {a**c}")
```

WTH is that wkwk

okok, 1e-05 tuh `0.00001` kok bisa? jadi :

$$10^{-5} = \frac{1}{10^5}$$

$$10^5 = 10 \times 10 \times 10 \times 10 \times 10 = 100.000$$

$$\frac{1}{100.000} = \mathbf{0.00001}$$

Untuk pangkat 10 negatif, cara tercepat adalah:

Angka $-5$ menunjukkan bahwa akan ada **5 angka nol** di depan angka 1, **termasuk** nol di depan koma (titik desimal).

$$\text{5 angka nol} \rightarrow 0.\underbrace{0000}_{4 \text{ nol}}1$$

Ini adalah alasan matematisnya mengapa $10^{-5}$ sama dengan $0.00001$.

## Operasi Modulus atau sisa bagi (%)

Jadi kita tahu kalo diatas tadi kita ngitung $10:3 = 3.3$. Nah kalo modulus tuh cuman ngasilin hasil sisanya. 

$10:3$ = kan $3$, sisa $1$. 

Jadi kalo $10\%3 = 1$.

```python
print(f"hasil dari {a}%{b} = {a%b}")
```

```python
print(f"hasil dari {a}%{c} = {a%c}")
```

*tau kan kenapa hasilnya 0?* karena $10:5$ dapet $2$, ga ada sisa, jadinya $10\%5 = 0$

```python
print(f"hasil dari {c}%{d} = {c%d}")
```

$$\frac{-5}{1} = -5 \text{ tanpa sisa } (0)$$

## Operasi Floor Division (//)

kalo Modulus kan yang ditampilin hasil sisanya, nah kalo floor division tuh yang ditampilin dapet berapanya.

hmm mungkin gini , 
- tadi kan pembagian $10:3=3.3$,
- trus kalo di modulus $10\%3=3$ karena sisa dari $10:3$ tuh $3$ yang kita ambil belakangnya $3.\`3\`$
- nah kalo di floor division, kita ambil dapetnya berapa, $10:3$ dapet `3`, jadi sisanya ga di itung. yang kita ambil depannya $\`3\`.3$

gini deh, kita ganti jadi 11 aja 

- pembagian $11:3=3.6$,
- sedangkan modulus $11\%3=2$ karena $11:3$ dapet $3$ asalnya $(3\times3=9)$ nah $11 - 9$ sisa $2$.
- kalo floor division, kita ambil dapetnya aja berapa? tadi diatas kita udah itung $11:3$ dapetnya $3$, yaudah berhenti disitu.

```python
print(11/3)
```

```python
print(11%3)
```

```python
print(11//3)
```
