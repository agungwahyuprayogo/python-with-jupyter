# Variable

variable adalah tempat penyimpanan data

sebagai permulaan kita akan menyimpan angka 10 di variable a

`````python
x = 2
`````

nah maksud dari :

```python
x = 2
````

adalah, kita meminta python untuk memasukan angka `2` ke variable `x`. ilustrasinya seperti ini :



kita coba langsung ya untuk membuktikan

`````python
x = 2
`````

`````python
y = x + 4
print(y)
`````



- `x` diatas sebutanya adalah **`variable`**, jadi tempat untuk menampung nilai
- `2` adalah **`value`** , angka atau nilai yang ingin kita masukan dalam variable
- dan `=` adalah **`assignmet`**. mungkin bisa dibilang kaya penjebatan yang ngasih tau ke python tolong masukin `2` ke `x`

`````python
a = 10
x = 77
panjang = 1001
`````

`````python
# pemanggilan variable
print(a)
print(x)
print(panjang)
`````

`````python
# bisa juga kita tambahkan kata - kata
print("Nilai a = ", a)
print("Nilai x = ", x)
print("Nilai panjang = ", panjang)
`````

Nah yang perlu diperhatikan dalam penamaan sebuah variable adalah jangan melakukan hal seperti ini :

`````python
nilai y = 10 # salah, karena nama variable mengandung spasi
`````

muncul 

```python
SyntaxError: invalid syntax
```

kenapa? karena penamaan **variable tidak boleh mengandung spasi** seperti `nilai y` diatas. 

*Lalu solusinya bagaimana?*

Bisa kita berikan tanda *`underscore`* atau `_` , maka jadinya **`nilai y`** menjadi -> **`nilai_y`**

`````python
nilai_y = 10 # betul, sudah tidak mengandung spasi
`````

Maka tidak akan terjadi error lagi, 

Lalu yang kedua, **jangan menempatkan angka sebagai awal nama sebuah variable**, contohnya seperti ini :

`````python
10juta = 10000000 # error, ada angka diawal variable
`````

Maka, lagi-lagi akan memunculkan `SyntaxError: invalid syntax`

*Lalu solusinya bagaimana?* kita bisa **pindahkan angka ke belakang**, contohnya **dari `10juta`** menjadi **`juta10`**

`````python
juta10 = 10000000
`````

`````python
juta-10 = 10000000 # error
`````

penggunaan tanda baca min **`-`** juga lebih baik dihindari saja, **python akan membaca `variable dikurangi 10`**, bukan memasukan angka `10000000` ke variable `juta-10`.

jadi, **ganti nama variable juta-10** menjadi **juta_10** atau **juta10**. 

`````python
juta_10 = 10000000
`````

---

selain `-`, tanda `+` seharusnya juga tidak bisa

`````python
juta+10 = 10000000
`````

ternyata betul. **selain spasi, disarankan untuk tidak memasukan operator matematika dalam penamaan variable**
