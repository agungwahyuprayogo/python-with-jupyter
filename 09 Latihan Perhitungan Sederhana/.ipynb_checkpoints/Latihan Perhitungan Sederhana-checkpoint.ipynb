# Latihan konversi satuan temperatur


## Tabel Lengkap Konversi Suhu

Berikut adalah tabel versi digital dari catatan tangan Anda, dengan bagian yang kosong sudah dilengkapi:

| Dari \ Ke | Celcius ($C$) | Reamur ($R$) | Fahrenheit ($F$) | Kelvin ($K$) |
| :--- | :---: | :---: | :---: | :---: |
| **Celcius** | — | $\frac{4}{5}C$ | $\frac{9}{5}C + 32$ | $C + 273$ |
| **Reamur** | $\frac{5}{4}R$ | — | $\frac{9}{4}R + 32$ | $\frac{5}{4}R + 273$ |
| **Fahrenheit**| $\frac{5}{9}(F - 32)$ | $\frac{4}{9}(F - 32)$ | — | **$\frac{5}{9}(F - 32) + 273$** |
| **Kelvin** | $K - 273$ | $\frac{4}{5}(K - 273)$ | **$\frac{9}{5}(K - 273) + 32$**| — |

### Tips Tambahan:
* **Reamur:** Skala ini jarang digunakan secara internasional, tetapi sering diajarkan di sekolah di Indonesia. Titik bekunya 0° dan titik didihnya 80°.
* **Kelvin:** Dalam sains yang sangat presisi, angka penambahnya sebenarnya adalah **273,15**. Namun, untuk pelajaran sekolah (SD/SMP/SMA), membulatkannya menjadi **273** 

## Celcius to others

Biar ga kebanyakan ngetik, kita masukan `input` angka aja ya, lalu kita bikin satu cell.

Semisal nya :
- *30°Celcius jadi berapa ke suhu lain?* 
- *oh hasilnya .. Reamur, .. Fahrenhait, dan .. Kelvin.*

Gausah kita code satu-satu celcius ke reamur, celcius ke Fahrenheit, celcius... dan seterusnya.

```python
awal_celcius = float(input("Masukan suhu awal celcius :"))
print(f"Suhu awal celcius yang Anda masukan adalah ➜ {awal_celcius}° Celcius")

print("\n")

# Celcsius to Reamur = 4/5 C
celcius_to_reamur = (4/5) * awal_celcius
print(f"{awal_celcius}° Celcius = {celcius_to_reamur}° Reamur")

# Celcsius to Fahrenheit = 9/5 C + 32
celcius_to_fahrenhait = ((9/5) * awal_celcius) + 32
print(f"{awal_celcius}° Celcius = {celcius_to_fahrenhait}° Fahrenhait")

# Celcsius to Kelvin = C + 273
celcius_to_kelvin = awal_celcius + 273
print(f"{awal_celcius}° Celcius = {celcius_to_kelvin}° Kelvin")
```

## Reamur to others

```python
awal_reamur = float(input("Masukan suhu awal reamur :"))
print(f"Suhu awal reamur yang Anda masukan adalah ➜ {awal_reamur}° Reamur")

print("\n")

# Reamur to Celcius = 5/4 R
reamur_to_celcius = (5/4) * awal_reamur
print(f"{awal_reamur}° Reamur = {reamur_to_celcius}° Celcius")

# Reamur to Fahrenhait = 9/4 R + 32
reamur_to_fahrenhait = ( (9/4) * awal_reamur) + 32
print(f"{awal_reamur}° Reamur = {reamur_to_fahrenhait}° Fahrenhait")

# Reamur to Kelvin = 5/4 R + 32
reamur_to_kelvin = ( (5/4) * awal_reamur) + 273
print(f"{awal_reamur}° Reamur = {reamur_to_kelvin}° Kelvin")
```

## Fahrenhait to others

```python
awal_fahrenhait = float(input("Masukan suhu awal Fahrenhait :"))
print(f"Suhu awal Fahrenhait yang Anda masukan adalah ➜ {awal_fahrenhait}° Fahrenhait")

print("\n")

# Fahrenhait to Celcius = 5/9 (F-32)
fahrenhait_to_celcius = (5/9) * (awal_fahrenhait - 32)
print(f"{awal_fahrenhait}° Fahrenhait = {fahrenhait_to_celcius}° Celcius")

# Fahrenhait to Reamur = 4/9 (F-32)
fahrenhait_to_reamur = (4/9) * (awal_fahrenhait - 32)
print(f"{awal_fahrenhait}° Fahrenhait = {fahrenhait_to_reamur}° Reamur")

# Fahrenhait to Kelvin = 5/9 (F-32) + 273
fahrenhait_to_kelvin = (5/9) * (awal_fahrenhait - 32) + 273
print(f"{awal_fahrenhait}° Fahrenhait = {fahrenhait_to_kelvin}° Kelvin")
```

## Kelvin to others

terakhir ya.. 

```python
awal_kelvin = float(input("Masukan suhu awal Kelvin :"))
print(f"Suhu awal Kelvin yang Anda masukan adalah ➜ {awal_kelvin}° Kelvin")

print("\n")

# Kelvin to Celcius = K - 273
kelvin_to_celcius = awal_kelvin - 273
print(f"{awal_kelvin}° kelvin = {kelvin_to_celcius}° Celcius")

# Kelvin to Reamur = (4/5) (K - 273)
kelvin_to_reamur = (4/5) * (awal_kelvin - 273)
print(f"{awal_kelvin}° kelvin = {kelvin_to_reamur}° Reamur")

# Kelvin to Fahrenhait = (9/5) (K - 273) + 32
kelvin_to_fahrenhait = (9/5) * (awal_kelvin - 273) + 32
print(f"{awal_kelvin}° kelvin = {kelvin_to_fahrenhait}° Fahrenhait")
```
