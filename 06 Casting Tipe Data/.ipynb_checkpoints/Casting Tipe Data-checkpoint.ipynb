# Casting Tipe Data
kita masuk ke `Casting Tipe Data` ya. Jadi kita akan latihan untuk **merubah satu tipe data ke tipe data lain**. 

Jadi seperti yang kita tahu, tipe data di Python sebelumnya ada `Integer`, `Float`, `String`, `Boolean`, `Complex`, dan beberapa tipe data dari bahasa pemograman C. 

Tapi kali ini **kita akan menggunakan tipe data yang sering digunakan di Python saja seperti `Integer`, `Float`, `String` dan `Boolean`**. Karena kecil kemungkinannya untuk casting data atau merubah tipe data dari atau ke tipe data complex maupun tipe data yang ada di bahasa C.

## Integer to others

Kita akan melihat perubahan data integer ke tipe data lain

````python
data_integer = 9 # ganti angka bebas
````

### Integer → Float, 

untuk mengubah ke tipe data `float` caranya cukup mudah :  

```python 
nama_variable = float(nama_variable)
```

````python
data_float = float(data_integer)
print(data_float)
````

oke, kita bisa ganti value di variable `data_integer` diatas dengan angka berapa pun selain koma. maka jika kita ubah menjadi float akan diberi `.0` dibelakang angka tersebut.

semisal kita memiliki : 

```python
data_integer = 9
```

jika kita casting ke float akan menjadi `9.0`

---

### Integer → String

untuk mengubah ke tipe data `string` caranya cukup mudah :  

```python 
nama_variable = str(nama_variable) # str sebagai string
```

````python
data_str = str(data_integer)
print(data_str)
````

oke, hasil print hanya menunjukan hasil `9`, alangkah baiknya jika kita lihat tipe datanya

````python
print(f"data_str : {data_str}, tipe data : {type(data_str)}")
````

oke, mungkin tidak akan terlihat secara sekilas perubahanya dari integer ke string, tapi untungnya kita bisa cek lagi menggunakan `type()`

---

### Integer → Boolean

untuk mengubah ke tipe data `boolean` caranya cukup mudah :  

```python 
nama_variable = boll(nama_variable) # bool sebagai boolean
```

````python
data_bool = bool(data_integer) 
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

karena hasil output dari tipe data `boolean` hanya dalam bentuk `True` atau `False`. Maka, kita bisa lakukan eksperimen dengan mengganti value dari variable `data_integer`

````python
data_integer = 1
data_bool = bool(data_integer) 
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

kita coba ganti angka lain

````python
data_integer = 9
data_bool = bool(data_integer) 
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

hasilnya masih `True`, kita ganti menjadi `0`

````python
data_integer = 0
data_bool = bool(data_integer) 
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

akhirnya terlihat perbedaan, hasilnya `False`. coba kita ganti ke angka minus

````python
data_integer = -1
data_bool = bool(data_integer) 
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

````python
data_integer = -9
data_bool = bool(data_integer) 
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

jadi.. **selama NILAI BUKAN `0` atau <> 0, hasilnya akan `True`**. Baru akan **bernilai False jika NILAI = 0**

---

## Float to others

Selanjutnya kita akan mencoba casting tipe data dari `float` ke `integer`, `string`, dan `boolean`.

````python
# kita inisiasi variable float diawal
data_float = 9.5
````

### Float → Integer

````python
data_int = int(data_float)
print(f"data_int : {data_int}, tipe data : {type(data_int)}")
````

dari `9.5` menjadi `9`. Coba kita ganti menjadi `9.9`, apakah akan menjadi 10 atau 9?

````python
data_float = 9.9
data_int = int(data_float)
print(f"data_int : {data_int}, tipe data : {type(data_int)}")
````

Ternyata hasilnya tetap 9. Jadi kesimpulannya adalah, **berapa pun angka dibelakang koma akan dihilangka.**

### Float → String

````python
data_str = str(data_float)
print(f"data_str : {data_str}, bertipe data : {type(data_str)}")
````

Mirip seperti `int to float`, tidak terlihat perubahan namun ketika kita cek tipe data baru terlihat jika tipe data sudah berubah menjadi string.

### Float → Boolean

````python
data_bool = bool(data_float)
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

kita tahu dari **integer ke boolean akan `False` jika `0`, selain `0` akan `True`**. *Apakah di float juga seperti itu?* kita coba

````python
data_float = 0.0
data_bool = bool(data_float)
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

````python
data_float = 0.1
data_bool = bool(data_float)
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

````python
data_float = 0.0000000000000000000000000000001
data_bool = bool(data_float)
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

okeh, kesimpulannya **harus 0.0 untuk float agar mendapatkan hasil False di boolean**

## String to others

kali ini kita coba untuk mengubah tipe data dari string ke tipe data lain. kita simpan sebuah variable awal terlebih dahulu :

````python
data_str = "Agung"
````

---

### String → Integer

````python
data_int = int(data_str)
````

Error ya ternyata. `ValueError: invalid literal for int() with base 10: 'Agung'` **Ya karena mana ada mengubah tipe data karakter (huruf) menjadi angka?** paling satu-satunya cara agar tidak error, *kita bisa mengganti `"Agung"` menjadi angka.*

````python
data_str = "10"
print(f"data_str : {data_str}, tipe data : {type(data_str)}")
````

````python
# conversi data to int
data_int = int(data_str)
print(f"data_int : {data_int}, tipe data : {type(data_int)}")
````

Kesimpulannya, dari **string ke integer baru berhasil jika nilai / value di dalam tanda kutip dalam bentuk angka**. Jika value string masih berupa kumpulan karakter, tidak akan bisa dikonversi menjadi integer. 

---

### String → Float

````python
# kita cek dulu data_str 
data_str
````

````python
data_float = float(data_str)
print(f"data_float : {data_float}, tipe data : {type(data_float)}")
````

Masih mirip seperti pengubahan tipe data apapun ke dalam bentuk float, akan selalu diberi tambahan `.0` Nah sekarang kita coba, apakah akan error juga jika kita mengganti tipe data string awal menggunakan kumpulan karakter.

````python
data_str = "Agung"
data_float = float(data_str)
print(f"data_float : {data_float}, tipe data : {type(data_float)}")
````

Mirip kejadian di integer, jika masih berupa kumpulan karakter atau huruf, akan terjadi `ValueError`

## Boolean to Others

oke, yang terakhir kali ini kita akan mencoba dari boolean ke tipe data lain. Kita buat variable baru dulu :

````python
data_bool = True
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
````

### Boolean → Integer

````python
data_int = int(data_bool)
print(f"data_int : {data_int}, tipe data : {type(data_int)}")
````

Hasilnya `True`, coba kita ganti value True menjadi `False`. Apakah akan berubah di casting integer?

````python
data_bool = False
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
data_int = int(data_bool)
print(f"data_int : {data_int}, tipe data : {type(data_int)}")
````

okay, jadi tetep ya. di boolean hasilnya False jika tipe data sebelumnya berupa angka 0, dan juga sebaliknya.

---

### Boolean → Float

````python
data_bool
````

Langsung saja ya, kita buat 2 code dimana 1 code kita isi value `False` dan `True`.

````python
data_bool = False
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
data_float = float(data_bool)
print(f"data_float : {data_float}, tipe data : {type(data_float)}")
````

````python
data_bool = True
print(f"data_bool : {data_bool}, tipe data : {type(data_bool)}")
data_float = float(data_bool)
print(f"data_float : {data_float}, tipe data : {type(data_float)}")
````

.. *well well*, dari Boolean ke Float juga sama seperti integer, dari **True ke Float menjadi 1.0**. dan **dari False ke Float menjadi 0.0**
