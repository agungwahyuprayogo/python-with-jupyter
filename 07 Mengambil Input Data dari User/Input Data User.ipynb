## Boolean

Nah kalo udah sampe Boolean nih kadang suka rada-rada

````python
boolean = bool(input("Masukan Boolean : "))   # hanya True or Falseb
````

````python
boolean
````

````python
boolean = bool(input("Masukan Boolean : "))   # hanya True or False
````

````python
boolean
````

Tuh kan! kita input False tapi pas di print lagi tuh hasilnya tetep True. 

Dari apa yang udah kita pelajarin kemarin, boolean baru bernilai False kalo.. angka inputnya 0. 

Nah jadi untuk mengatasi itu, kita perlu **casting input data user ke *`Integer` terlebih dahulu, baru ke `Boolean`***

````python
boolean = bool(int(input("Masukan Angka Boolean :")))
print(boolean)
````

````python
boolean = bool(int(input("Masukan Angka Boolean :")))
print(boolean)
````

Nahh.. jadi seperti itu mengatasi Boolean, kita perlu casting ke Integer terlebih dahulu baru casting lagi ke boolean
